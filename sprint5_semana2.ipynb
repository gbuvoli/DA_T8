## 1) üîé Filtrar datasets (selecci√≥n de datos)
Filtrar es **quedarte con una parte del dataset** seg√∫n un criterio. Se usa para:
- analizar un segmento espec√≠fico (por ejemplo, un pa√≠s, un producto, una cohorte),
- excluir registros inv√°lidos o fuera de alcance,
- comparar grupos (A vs B) usando reglas consistentes.

Conceptos clave:
- **Filtrado por condiciones**: seleccionar filas que cumplan una regla (igualdad, rangos, pertenencia a una lista, combinaciones).
- **Combinaci√≥n de condiciones**: unir reglas con l√≥gica (y / o / no).
- **Filtrado por columnas**: quedarte solo con las variables necesarias para el an√°lisis (reduce ruido y errores).
- **Filtrado por fechas**: seleccionar por intervalos temporales; requiere que las fechas est√©n bien interpretadas como fechas (no como texto).
- **Evitar ‚Äúfiltros que mienten‚Äù**: si hay nulos, categor√≠as inconsistentes o tipos errados, el filtro puede excluir o incluir mal (por eso se explora y limpia antes).

Resultado esperado: un dataset m√°s peque√±o y relevante, listo para calcular m√©tricas o construir tablas resumen.

---

```python
# Filtro por una condici√≥n

adultos = df[df["Age"] >= 18]
display(adultos.head())
```

```python
# Filtro multiples condiciones

fem_adultos = df[(df["Age"] >= 18) & (df["Sex"] == "female")]
display(fem_adultos.head())
```

```python
#Filtrar por lista de valores

clase_1_2 = df[df["Pclass"].isin([1, 2])]
display(clase_1_2.head())
```

```python
#Filtrar por valores nulos
nulos_age = df[df["Age"].isnull()]
display(nulos_age.head())
```

```python
#Filtrar por rango de edad
rango_edad = df[df["Age"].between(30, 40)]
display(rango_edad.head())
```

```python
# Filtrar por filas y columnas espec√≠ficas
subset = df.loc[df["Age"] >= 18, ["Name", "Age", "Sex"]]
display(subset.head())
```



## 2) üß© Agrupar datos (res√∫menes por categor√≠a)
Agrupar es convertir datos ‚Äúa nivel registro‚Äù en **informaci√≥n agregada** por una o varias categor√≠as (por ejemplo, por ciudad, mes, canal, tipo de cliente). Es esencial para responder preguntas de negocio.

Conceptos clave:
- **Clave(s) de agrupaci√≥n**: columnas que definen los grupos (una o varias).  
  Ejemplo conceptual: agrupar por `ciudad` y `mes` crea un grupo por cada combinaci√≥n ciudad‚Äìmes.
- **Agregaciones**: operaciones que resumen cada grupo (conteo, suma, promedio, mediana, m√≠nimo/m√°ximo, porcentajes, etc.).
- **M√©tricas derivadas**: despu√©s de agrupar, a menudo se calculan ratios (por ejemplo, tasa, margen, participaci√≥n, crecimiento).
- **Orden y legibilidad**: un buen resumen suele incluir ordenamiento por una m√©trica clave y un formato claro.
- **Granularidad**: agrupar demasiado fino (muchas categor√≠as) produce ruido; agrupar demasiado grueso oculta patrones. La clave es elegir el nivel correcto para la pregunta.

Resultado esperado: una tabla resumen que responde ‚Äúqu√© pasa‚Äù por segmento.


![alt text](https://pbpython.com/images/groupby-example.png)

![alt text](https://keytodatascience.com/wp-content/uploads/2020/04/image-1.png)

```python
#Supervivencia
survival_rate = df.groupby(['Survived'])['PassengerId'].count().reset_index()
display(survival_rate)
```

```python
#Visualizaci√≥n de la tasa de supervivencia
survival_rate.plot(kind='bar', x='Survived', y='PassengerId', title='Survival Rate', legend=False, ylabel='Number of Passengers', xlabel='Survived', color=['red', 'green'])
```

```python
#survival age

survival_age = df.groupby(['Survived'])['Age'].mean().reset_index()
display(survival_age)
```

```python
#plot
survival_age.plot(kind='bar', x='Survived', y='Age', title='Average Age by Survival Status', legend=False, ylabel='Average Age', xlabel='Survived', color=['orange', 'blue'])
```

```python
# survival avg age and avg fare

survival_stats = df.groupby(['Survived'])[['Age', 'Fare']].mean().reset_index()
#ajuste de nombres de columnas
survival_stats.columns = ['Survived', 'Avg_Age', 'Avg_Fare']
display(survival_stats)
```

```python
#visualizaci√≥n
survival_stats.plot(kind='bar', x='Survived', y=['Avg_Age', 'Avg_Fare'], title='Average Age and Fare by Survival Status', ylabel='Values', xlabel='Survived', color=['orange', 'blue'])
```

```python
#survival avg age and sum fare

survival_stats_sum = df.groupby(['Survived'])[['Age', 'Fare']].agg(

    Avg_Age = ('Age', 'mean'),
    Total_Fare = ('Fare', 'sum')
).reset_index()

display(survival_stats_sum)
```

```python
#visualizaci√≥n
survival_stats_sum.plot(kind='bar', x='Survived', y=['Avg_Age', 'Total_Fare'], title='Average Age and Fare by Survival Status', ylabel='Values', xlabel='Survived', color=['orange', 'blue'])
```

```python
# Edad promedio por clase y supervivencia
survival_by_class = df.groupby(['Pclass', 'Survived'])['Age'].mean()
display(survival_by_class)
```

```python
survival_by_class.plot(kind='bar',y='Age', title='Average Age by Class and Survival Status', ylabel='Average Age', xlabel='Passenger Class')
```

```python
survival_by_class = df.groupby(['Pclass', 'Survived'])['Age'].mean().unstack()

display(survival_by_class)
```

```python
survival_by_class.plot(kind='bar', title='Average Age by Class and Survival Status', ylabel='Average Age', xlabel='Passenger Class', color=['orange', 'blue'])
```
